```python
import os
import glob
import subprocess

# ----------------------
# Rutas
# ----------------------
INPUT_FAST5 = "/content/drive/MyDrive/input_files"
RESULTS_DIR = "/content/drive/MyDrive/results/CPU"
POD5_CONVERT = "/content/drive/MyDrive/apps/pod5-0.3.34/src/pod5/tools/pod5_convert_from_fast5.py"
POD5_MERGE = "/content/drive/MyDrive/apps/pod5-0.3.34/src/pod5/tools/pod5_merge.py"
LOG_FILE = os.path.join(RESULTS_DIR, "conversion_errors.log")

os.makedirs(RESULTS_DIR, exist_ok=True)

# Crear / limpiar log
with open(LOG_FILE, "w") as log:
    log.write("=== LOG DE ERRORES DE CONVERSIÓN FAST5 → POD5 ===\n\n")

# ----------------------
# Procesar cada carpeta/barcode
# ----------------------
for folder in sorted(os.listdir(INPUT_FAST5)):
    folder_path = os.path.join(INPUT_FAST5, folder)
    if not os.path.isdir(folder_path):
        continue
    print(f"\n🧬 Procesando carpeta: {folder}")

    pod5_folder = os.path.join(RESULTS_DIR, folder, "T-POD5")
    os.makedirs(pod5_folder, exist_ok=True)

    fast5_files = glob.glob(os.path.join(folder_path, "*.fast5"))
    if not fast5_files:
        print(f"⚠️ No se encontraron archivos .fast5 en {folder_path}")
        continue

    pod5_files = []

    # ----------------------
    # Convertir FAST5 → POD5
    # ----------------------
    for fast5_file in fast5_files:
        pod5_out = os.path.join(pod5_folder, os.path.basename(fast5_file).replace(".fast5", ".pod5"))

        # Saltar si ya existe
        if os.path.exists(pod5_out):
            print(f"⏩ Ya existe: {pod5_out}")
            pod5_files.append(pod5_out)
            continue

        print(f"🔄 Convirtiendo: {os.path.basename(fast5_file)}")

        result = subprocess.run([
            "python3",
            POD5_CONVERT,
            fast5_file,
            "-o", pod5_out,
            "-t", "4"
        ], capture_output=True, text=True)

        if result.returncode != 0:
            print(f"❌ Error convirtiendo {fast5_file}")
            print("↳ Detalles guardados en el log.")
            with open(LOG_FILE, "a") as log:
                log.write(f"\n--- ERROR EN: {fast5_file} ---\n")
                log.write(f"STDOUT:\n{result.stdout}\n")
                log.write(f"STDERR:\n{result.stderr}\n")
                log.write("-" * 60 + "\n")
            continue  # Saltar este archivo

        pod5_files.append(pod5_out)
        print(f"✅ Convertido correctamente: {os.path.basename(fast5_file)}")

    # ----------------------
    # Merge de todos los POD5 del barcode
    # ----------------------
    if not pod5_files:
        print(f"⚠️ No se generaron archivos POD5 válidos para {folder}.")
        continue

    merged_pod5 = os.path.join(pod5_folder, f"{folder}.pod5")
    print(f"🧩 Uniendo {len(pod5_files)} archivos POD5 → {merged_pod5}")

    merge_result = subprocess.run([
        "python3",
        POD5_MERGE,
        "-o", merged_pod5
    ] + pod5_files, capture_output=True, text=True)

    if merge_result.returncode != 0:
        print(f"❌ Error al hacer merge en {folder}")
        with open(LOG_FILE, "a") as log:
            log.write(f"\n--- ERROR EN MERGE: {folder} ---\n")
            log.write(f"STDOUT:\n{merge_result.stdout}\n")
            log.write(f"STDERR:\n{merge_result.stderr}\n")
            log.write("-" * 60 + "\n")
    else:
        print(f"✅ Archivo POD5 final generado: {merged_pod5}")

        # ----------------------
        # 🔹 Eliminar archivos POD5 intermedios
        # ----------------------
        print("🧹 Eliminando archivos POD5 intermedios...")
        for f in pod5_files:
            if os.path.exists(f) and f != merged_pod5:
                try:
                    os.remove(f)
                except Exception as e:
                    print(f"[WARN] No se pudo eliminar {f}: {e}")
        print("✅ Archivos intermedios eliminados correctamente.")
```

```python
import os
import subprocess
import time
import psutil
import csv

# ================================
# CONFIGURACIÓN GENERAL
# ================================
base_dir = "/content/drive/MyDrive"
dorado_bin = f"{base_dir}/apps/dorado-0.9.6-linux-x64/bin/dorado"
model_path = f"{base_dir}/apps/dorado-0.9.6-linux-x64/models/dna_r9.4.1_e8_fast@v3.4"
results_dir = f"{base_dir}/results/CPU"

# Archivos donde guardaremos métricas
sysmon_csv = "/content/dorado_sysmon.csv"
runtime_csv = "/content/dorado_runtime.csv"

# Crear CSV si no existen
if not os.path.exists(sysmon_csv):
    with open(sysmon_csv, "w") as f:
        writer = csv.writer(f)
        writer.writerow(["timestamp", "barcode", "cpu_percent", "ram_mb",
                         "gpu_util", "gpu_mem", "elapsed_s"])

if not os.path.exists(runtime_csv):
    with open(runtime_csv, "w") as f:
        writer = csv.writer(f)
        writer.writerow(["barcode", "exit_code", "time_seconds", "success"])


def get_gpu_stats():
    """Devuelve GPU usage si existe, sino None."""
    try:
        result = subprocess.check_output(
            "nvidia-smi --query-gpu=utilization.gpu,memory.used "
            "--format=csv,noheader,nounits", shell=True
        ).decode().strip()
        util, mem = result.split(", ")
        return int(util), int(mem)
    except:
        return None, None


# ================================
# LOOP SOBRE LOS BARCODES
# ================================
print("🔁 Procesando todos los barcodes...\n")

for barcode_folder in os.listdir(results_dir):
    barcode_path = os.path.join(results_dir, barcode_folder)
    t_pod5_path = os.path.join(barcode_path, "T-POD5")
    pod5_file = os.path.join(t_pod5_path, f"{barcode_folder}.pod5")

    if not os.path.isdir(t_pod5_path) or not os.path.isfile(pod5_file):
        print(f"[⚠️] No se encontró archivo .pod5 en {barcode_folder}, saltando...\n")
        continue

    # Carpeta de salida para Dorado
    output_dir = os.path.join(barcode_path, "BASECALLED")
    os.makedirs(output_dir, exist_ok=True)

    # Archivo log
    dorado_log = os.path.join(output_dir, f"{barcode_folder}_dorado.log")

    print(f"[{barcode_folder}] Procesando {os.path.basename(pod5_file)}...")

    # Comando Dorado
    cmd = [
        dorado_bin, "basecaller",
        model_path,
        pod5_file,
        "--emit-fastq",
        "--min-qscore", "10",
        "--output-dir", output_dir,
        "--verbose"
    ]

    start_time = time.time()

    # =============================
    # EJECUCIÓN + CAPTURA DE LOG
    # =============================
    with open(dorado_log, "w") as logf:
        process = subprocess.Popen(
            cmd,
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            text=True,
            bufsize=1
        )

        # =============================
        # CAPTURA DE SISTEMA EN TIEMPO REAL
        # =============================
        while True:
            retcode = process.poll()

            # Leer logs sin bloquear
            line = process.stdout.readline()
            if line:
                logf.write(line)

            # Guardar métricas cada segundo
            elapsed = time.time() - start_time
            cpu = psutil.cpu_percent(interval=None)
            ram = psutil.virtual_memory().used / (1024 * 1024)
            gpu_util, gpu_mem = get_gpu_stats()

            with open(sysmon_csv, "a") as f:
                writer = csv.writer(f)
                writer.writerow([
                    time.time(), barcode_folder, cpu,
                    int(ram), gpu_util, gpu_mem, round(elapsed, 2)
                ])

            if retcode is not None:
                break
            time.sleep(1)

    end_time = time.time()
    total_time = round(end_time - start_time, 2)

    # Guardar registro de resultado del proceso
    with open(runtime_csv, "a") as f:
        writer = csv.writer(f)
        writer.writerow([barcode_folder, retcode, total_time, retcode == 0])

    print(f"[⏱] Tiempo total: {total_time} s")

    # =============================
    # Renombrar FASTQ generado
    # =============================
    fastqs = [f for f in os.listdir(output_dir) if f.endswith(".fastq")]
    if fastqs:
        old_path = os.path.join(output_dir, fastqs[0])
        new_path = os.path.join(output_dir, f"{barcode_folder}.fastq")
        os.rename(old_path, new_path)
        print(f"[📄] Archivo renombrado a: {barcode_folder}.fastq\n")
    else:
        print(f"[⚠️] No se generó archivo FASTQ en {output_dir}\n")

print("🎯 Procesamiento completo.")
```

```python
# ==============================================================
# 🚀 TRIMMING AUTOMÁTICO CON PORECHOP_ABI EN GOOGLE COLAB
# ==============================================================

import os
import subprocess
import time
from tqdm import tqdm

# -------------------------------------------------------------
# CONFIGURACIÓN
# -------------------------------------------------------------
BASE_DIR = "/content/drive/MyDrive/results/GPU"
PORECHOP_BIN = "/usr/local/miniforge3/envs/porechop_env/bin/porechop_abi"
THREADS = 2

os.makedirs(BASE_DIR, exist_ok=True)
print(f"📂 Carpeta base: {BASE_DIR}")
print(f"🧩 Binario de Porechop: {PORECHOP_BIN}")

# -------------------------------------------------------------
# FUNCIONES AUXILIARES
# -------------------------------------------------------------
def count_reads(fastq_path):
    """Cuenta lecturas en un archivo FASTQ."""
    with open(fastq_path, "r") as f:
        lines = sum(1 for _ in f)
    return lines // 4

def run_porechop(fastq_path, output_path, log_path):
    """Ejecuta porechop_abi con barra de progreso."""
    total_reads = count_reads(fastq_path)
    cmd = [
        PORECHOP_BIN,
        "-i", fastq_path,
        "-o", output_path,
        "--threads", str(THREADS),
        "--discard_middle"
    ]

    with open(log_path, "w") as log_file:
        process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)

        pbar = tqdm(total=total_reads, desc="🔬 Trimming", unit="reads", ncols=80)
        reads_processed = 0

        for line in process.stdout:
            log_file.write(line)
            log_file.flush()
            if "Processing read" in line or "Trimmed" in line:
                reads_processed += 1
                if reads_processed < total_reads:
                    pbar.update(1)

        process.wait()
        pbar.close()
        return process.returncode

# -------------------------------------------------------------
# PROCESAMIENTO PRINCIPAL
# -------------------------------------------------------------
for barcode_dir in sorted(os.listdir(BASE_DIR)):
    barcode_path = os.path.join(BASE_DIR, barcode_dir)
    if not os.path.isdir(barcode_path):
        continue

    basecalled_dir = os.path.join(barcode_path, "BASECALLED")
    if not os.path.exists(basecalled_dir):
        print(f"⚠️  No se encontró BASECALLED en {barcode_dir}")
        continue

    fastq_files = [f for f in os.listdir(basecalled_dir) if f.endswith(".fastq")]
    if not fastq_files:
        print(f"⚠️  No hay archivos FASTQ en {barcode_dir}")
        continue

    fastq_path = os.path.join(basecalled_dir, fastq_files[0])
    porechop_dir = os.path.join(barcode_path, "PORECHOP")
    os.makedirs(porechop_dir, exist_ok=True)

    output_path = os.path.join(porechop_dir, fastq_files[0].replace(".fastq", "_trimmed.fastq"))
    log_path = os.path.join(porechop_dir, "porechop.log")

    print(f"\n🔹 Procesando {barcode_dir}...")
    start_time = time.time()

    code = run_porechop(fastq_path, output_path, log_path)

    if code == 0:
        elapsed = time.time() - start_time
        print(f"✅ {barcode_dir} completado en {elapsed/60:.2f} min. Resultado: {output_path}")
    else:
        print(f"❌ Error en {barcode_dir}. Revisa el log: {log_path}")

print("\n🎯 Trimming finalizado para todos los barcodes.")
```

```python
#!/usr/bin/env python3
"""
hybrid_tax_pipeline_v3_racon_full.py

Pipeline híbrido BLAST + SINTAX con:
- Polishing iterativo por RACON
- Overlaps generados con minimap2
- Carpetas RACON separadas de VSEARCH
- Logging detallado por barcode y por ronda
- Fallback automático a consenso VSEARCH
- Reglas A–C (Regla D eliminada)
"""

import os
import glob
import re
import math
import subprocess
import logging
import multiprocessing as mp
from collections import Counter

import pandas as pd
import numpy as np
from Bio import SeqIO

import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

# ================= CONFIG =================
BASE_DIR = "/content/drive/MyDrive/results/CPU"
WORKDIR = os.path.join(BASE_DIR, "work")
os.makedirs(WORKDIR, exist_ok=True)

# ---- binaries ----
RACON_BIN = "/content/drive/MyDrive/apps/racon/build/bin/racon"
MINIMAP2_BIN = "/content/drive/MyDrive/apps/minimap2-2.30_x64-linux/minimap2"
BLASTN_BIN = "/content/drive/MyDrive/apps/ncbi-blast-2.17.0+/bin/blastn"
VSEARCH_BIN = "/content/drive/MyDrive/apps/vsearch-2.23.0-linux-x86_64/bin/vsearch"

# ---- databases ----
BLAST_DB_PREFIX = "/content/drive/MyDrive/logs/blastdb/unite2024ITS"
UNITE_FASTA = "/content/drive/MyDrive/logs/unite/UNITE_public_19.02.2025.fasta"
SINTAX_FASTA = "/content/drive/MyDrive/logs/unite/utax_reference_dataset_19.02.2025.fasta"

# ---- files ----
OTU_QUERIES = os.path.join(WORKDIR, "otu_queries.fasta")
BLAST_OUT = os.path.join(WORKDIR, "blast_results.tsv")
SINTAX_OUT = os.path.join(WORKDIR, "sintax_results.tsv")
FINAL_ONE = os.path.join(WORKDIR, "final_taxonomy.tsv")
FINAL_VERBOSE = os.path.join(WORKDIR, "final_taxonomy_verbose.tsv")
UNASSIGNED = os.path.join(WORKDIR, "unassigned.tsv")

PLOTS_DIR = os.path.join(WORKDIR, "plots")
os.makedirs(PLOTS_DIR, exist_ok=True)

# ---- RACON ----
RACON_ROUNDS = 4

# ---- thresholds ----
SINTAX_SPECIES_CUTOFF = 0.90
SINTAX_GENUS_CUTOFF = 0.90
SPECIES_PID = 97.0
SPECIES_QCOV = 90.0
GENUS_PID = 94.0
GENUS_QCOV = 80.0
RULE_B_PID = 99.0
RULE_B_QCOV = 95.0
DELTA_PID_FOR_B = 1.0

BLAST_MAX_TARGETS = 10
BLAST_THREADS = 4
BLAST_EVALUE = 1e-5
PARALLEL_THRESHOLD = 500

# ================= LOGGING =================
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(message)s",
    handlers=[logging.StreamHandler()]
)
logger = logging.getLogger("hybrid_pipeline")

# ================= UTILITIES =================
def run_cmd(cmd):
    p = subprocess.run(
        cmd,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )
    return p.returncode, p.stdout, p.stderr

def safe_float(x, default=0.0):
    try:
        return float(x)
    except Exception:
        return default

def normalize_name(name):
    if not name or pd.isna(name):
        return ""
    return str(name).replace("_", " ").strip().lower()

def compute_penalty(identity, coverage):
    try:
        return math.log(1 + (100 - identity)) + math.log(1 + (100 - coverage))
    except Exception:
        return 0.0

# ================= RACON POLISHING =================
def run_racon_for_barcode(barcode_dir):
    barcode = os.path.basename(barcode_dir)
    logger.info("🔧 [Racon] Processing barcode %s", barcode)

    # -------- inputs --------
    reads_fastq = glob.glob(
        os.path.join(barcode_dir, "PORECHOP", "*_trimmed.fastq")
    )
    consensus0 = os.path.join(
        barcode_dir, "VSEARCH", "polished_best.fasta"
    )

    if not reads_fastq or not os.path.exists(consensus0):
        logger.warning(
            "⚠️ [Racon] Missing reads or consensus for %s → using VSEARCH consensus",
            barcode
        )
        return consensus0

    reads_fastq = reads_fastq[0]

    racon_dir = os.path.join(barcode_dir, "RACON")
    os.makedirs(racon_dir, exist_ok=True)

    # -------- convert reads --------
    reads_fasta = os.path.join(racon_dir, "reads.fasta")
    if not os.path.exists(reads_fasta):
        SeqIO.convert(reads_fastq, "fastq", reads_fasta, "fasta")

    prev_consensus = consensus0

    for r in range(1, RACON_ROUNDS + 1):
        logger.info(
            "🔁 [Racon] Barcode %s — round %d / %d",
            barcode, r, RACON_ROUNDS
        )

        overlaps = os.path.join(racon_dir, f"overlaps_round{r}.paf")
        out_fa = os.path.join(racon_dir, f"racon_round{r}.fasta")

        # -------- generate overlaps --------
        logger.info(
            "🧬 [minimap2] Barcode %s — generating overlaps (round %d)",
            barcode, r
        )

        ret, out, err = run_cmd([
            MINIMAP2_BIN,
            "-x", "map-ont",
            prev_consensus,
            reads_fasta
        ])

        if ret != 0 or not out.strip():
            logger.warning(
                "⚠️ [minimap2] No overlaps for %s at round %d → stopping",
                barcode, r
            )
            break

        with open(overlaps, "w") as f:
            f.write(out)

        # -------- racon --------
        ret, racon_out, err = run_cmd([
            RACON_BIN,
            reads_fasta,
            overlaps,
            prev_consensus
        ])

        if ret != 0 or not racon_out.strip():
            logger.warning(
                "⚠️ [Racon] Failed or empty output for %s at round %d → stopping",
                barcode, r
            )
            break

        with open(out_fa, "w") as f:
            f.write(racon_out)

        prev_consensus = out_fa

    final_fa = os.path.join(racon_dir, "racon_final.fasta")

    if prev_consensus != consensus0:
        os.replace(prev_consensus, final_fa)
        logger.info("✅ [Racon] Finished barcode %s", barcode)
        return final_fa
    else:
        logger.info(
            "ℹ️ [Racon] No improvement for %s → using VSEARCH consensus",
            barcode
        )
        return consensus0

    with open(overlaps, "w") as f:
        f.write(out)

    # iterative racon
    prev = consensus
    for r in range(1, RACON_ROUNDS + 1):
        out_fa = os.path.join(racon_dir, f"racon_round{r}.fasta")
        logger.info(
            "🔁 [Racon] Barcode %s — round %d / %d",
            barcode, r, RACON_ROUNDS
        )

        ret, out, err = run_cmd([
            RACON_BIN,
            reads_fasta,
            overlaps,
            prev
        ])

        if ret != 0:
            logger.warning(
                "⚠️ [Racon] Failed round %d for %s → fallback",
                r, barcode
            )
            return consensus

        with open(out_fa, "w") as f:
            f.write(out)

        prev = out_fa

    final_fa = os.path.join(racon_dir, "racon_final.fasta")
    os.rename(prev, final_fa)

    logger.info("✅ [Racon] Finished barcode %s", barcode)
    return final_fa

# ================= GATHER OTU QUERIES =================
def gather_queries():
    logger.info("📥 Gathering final consensus sequences")
    count = 0

    with open(OTU_QUERIES, "w") as out:
        for bc in sorted(glob.glob(os.path.join(BASE_DIR, "H*"))):
            final_consensus = run_racon_for_barcode(bc)

            if not os.path.exists(final_consensus):
                continue

            for rec in SeqIO.parse(final_consensus, "fasta"):
                rec.id = f"{os.path.basename(bc)}__{rec.id}"
                rec.description = ""
                SeqIO.write(rec, out, "fasta")
                count += 1

    if count == 0:
        raise SystemExit("❌ No OTUs collected")

    logger.info("✅ Collected %d OTUs for downstream analysis", count)
# ================= FASTA SPLITTING =================
def split_fasta(in_fasta, n_chunks, out_prefix):
    records = list(SeqIO.parse(in_fasta, "fasta"))
    if not records:
        return []
    chunk_size = max(1, int(math.ceil(len(records) / n_chunks)))
    out_files = []
    for i in range(0, len(records), chunk_size):
        chunk = records[i:i + chunk_size]
        out_fa = f"{out_prefix}.{i//chunk_size}.fasta"
        SeqIO.write(chunk, out_fa, "fasta")
        out_files.append(out_fa)
    return out_files

# ================= BLAST =================
def _blast_worker(query_fasta, out_path):
    cmd = [
        BLASTN_BIN,
        "-query", query_fasta,
        "-db", BLAST_DB_PREFIX,
        "-outfmt",
        "6 qseqid sseqid pident length mismatch gapopen qstart qend "
        "sstart send evalue bitscore qcovs qcovhsp stitle",
        "-max_target_seqs", str(BLAST_MAX_TARGETS),
        "-num_threads", "1",
        "-evalue", str(BLAST_EVALUE)
    ]
    ret, out, err = run_cmd(cmd)
    if ret != 0:
        raise RuntimeError(err)
    with open(out_path, "w") as f:
        f.write(out)
    return out_path

def run_blast(parallel=False):
    logger.info("🧬 Running BLAST (parallel=%s)", parallel)

    if not parallel:
        cmd = [
            BLASTN_BIN,
            "-query", OTU_QUERIES,
            "-db", BLAST_DB_PREFIX,
            "-outfmt",
            "6 qseqid sseqid pident length mismatch gapopen "
            "qstart qend sstart send evalue bitscore qcovs qcovhsp stitle",
            "-max_target_seqs", str(BLAST_MAX_TARGETS),
            "-num_threads", str(BLAST_THREADS),
            "-evalue", str(BLAST_EVALUE)
        ]

        ret, out, err = run_cmd(cmd)
        if ret != 0:
            raise RuntimeError(err)

        with open(BLAST_OUT, "w") as f:
            f.write(out)

    else:
        records = list(SeqIO.parse(OTU_QUERIES, "fasta"))
        n = len(records)

        n_chunks = min(mp.cpu_count(), max(1, n // 100))
        prefix = os.path.join(WORKDIR, "blast_chunk")
        chunks = split_fasta(OTU_QUERIES, n_chunks, prefix)

        pool = mp.Pool(processes=len(chunks))
        for c in chunks:
            pool.apply_async(_blast_worker, (c, c + ".blast.tsv"))

        pool.close()
        pool.join()

        with open(BLAST_OUT, "w") as out:
            for c in chunks:
                p = c + ".blast.tsv"
                if os.path.exists(p):
                    with open(p) as f:
                        out.write(f.read())

# ================= SINTAX =================
def _sintax_worker(query_fasta, out_path):
    cmd = [
        VSEARCH_BIN,
        "--sintax", query_fasta,
        "--db", SINTAX_FASTA,
        "--tabbedout", out_path,
        "--sintax_cutoff", "0.0",
        "--strand", "both",
        "--threads", "1"
    ]
    ret, _, err = run_cmd(cmd)
    if ret != 0:
        raise RuntimeError(err)
    return out_path

def run_sintax(parallel=False):
    logger.info("🧬 Running SINTAX (parallel=%s)", parallel)

    if not parallel:
        cmd = [
            VSEARCH_BIN,
            "--sintax", OTU_QUERIES,
            "--db", SINTAX_FASTA,
            "--tabbedout", SINTAX_OUT,
            "--sintax_cutoff", "0.0",
            "--strand", "both",
            "--threads", str(BLAST_THREADS)
        ]
        ret, _, err = run_cmd(cmd)
        if ret != 0:
            raise RuntimeError(err)
    else:
        records = list(SeqIO.parse(OTU_QUERIES, "fasta"))
        n = len(records)
        n_chunks = min(mp.cpu_count(), max(1, n // 200))
        prefix = os.path.join(WORKDIR, "sintax_chunk")
        chunks = split_fasta(OTU_QUERIES, n_chunks, prefix)

        pool = mp.Pool(processes=len(chunks))
        outs = []
        for c in chunks:
            outs.append(
                pool.apply_async(
                    _sintax_worker,
                    (c, c + ".sintax.tsv")
                )
            )
        pool.close()
        pool.join()

        with open(SINTAX_OUT, "w") as out:
            for c in chunks:
                p = c + ".sintax.tsv"
                if os.path.exists(p):
                    with open(p) as f:
                        out.write(f.read())

# ================= UNITE ACCESSION MAP =================
def build_unite_accession_map(unite_fasta):
    logger.info("📚 Building UNITE accession map")
    acc_map = {}
    with open(unite_fasta) as f:
        for line in f:
            if not line.startswith(">"):
                continue
            header = line.strip()
            genus = ""
            species = ""
            mg = re.search(r"g__([^;|]+)", header)
            ms = re.search(r"s__([^;|]+)", header)
            if mg:
                genus = mg.group(1)
            if ms:
                species = ms.group(1)

            tokens = re.split(r"[>\s|]+", header)
            for tok in tokens:
                if tok and tok not in acc_map:
                    acc_map[tok] = (genus, species)
    return acc_map

# ================= PARSE BLAST =================
def parse_blast(blast_out, acc_map):
    cols = [
        "qseqid","sseqid","pident","length","mismatch","gapopen",
        "qstart","qend","sstart","send","evalue","bitscore",
        "qcovs","qcovhsp","stitle"
    ]
    df = pd.read_csv(blast_out, sep="\t", names=cols)

    df["pident"] = pd.to_numeric(df["pident"], errors="coerce").fillna(0.0)
    df["qcovhsp"] = pd.to_numeric(df["qcovhsp"], errors="coerce").fillna(0.0)
    df["bitscore"] = pd.to_numeric(df["bitscore"], errors="coerce").fillna(0.0)

    df = df[df["qcovhsp"] > 1]

    def map_acc(a):
        if pd.isna(a):
            return ("","")
        if a in acc_map:
            return acc_map[a]
        base = a.split(".")[0]
        return acc_map.get(base, ("",""))

    df[["blast_genus","blast_species"]] = (
        df["sseqid"].apply(lambda x: pd.Series(map_acc(x)))
    )

    group = df.groupby("qseqid").agg(
        avg_pid=("pident","mean"),
        avg_qcov=("qcovhsp","mean"),
        n_hits=("sseqid","count"),
        best_bits=("bitscore","max")
    ).reset_index()

    df = df.merge(group, on="qseqid", how="left")
    return df

# ================= PARSE SINTAX =================
def parse_sintax(sintax_out):
    rows = []
    with open(sintax_out) as f:
        for line in f:
            parts = line.strip().split("\t")
            if len(parts) < 2:
                continue
            q, annot = parts[0], parts[1]
            genus = species = ""
            gscore = sscore = 0.0

            for tok in annot.split(","):
                tok = tok.strip()
                if tok.startswith("g:"):
                    m = re.match(r"g:([^(\s]+)(?:\(([\d\.]+)\))?", tok)
                    if m:
                        genus = m.group(1)
                        gscore = safe_float(m.group(2))
                if tok.startswith("s:"):
                    m = re.match(r"s:([^(\s]+)(?:\(([\d\.]+)\))?", tok)
                    if m:
                        species = m.group(1)
                        sscore = safe_float(m.group(2))

            rows.append((q, genus, gscore, species, sscore))

    return pd.DataFrame(
        rows,
        columns=[
            "qseqid",
            "sintax_genus",
            "sintax_genus_score",
            "sintax_species",
            "sintax_species_score"
        ]
    )
# ================= TAXONOMIC DECISION =================
def assign_taxonomy(df):
    assignments = []

    for q, sub in df.groupby("qseqid"):
        row = {
            "qseqid": q,
            "method": "unassigned",
            "genus": "",
            "species": "",
            "confidence": 0.0
        }

        # ---------- Rule A: BLAST species ----------
        blast_ok = sub[
            (sub["avg_pid"] >= SPECIES_PID) &
            (sub["avg_qcov"] >= SPECIES_QCOV) &
            (sub["blast_species"] != "")
        ]

        if not blast_ok.empty:
            best = blast_ok.sort_values("avg_pid", ascending=False).iloc[0]
            row.update({
                "rule": "A",
                "method": "BLAST_species",
                "genus": best["blast_genus"],
                "species": best["blast_species"],
                "confidence": best["avg_pid"]
            })
            assignments.append(row)
            continue

        # ---------- Rule B: BLAST genus (tight) ----------
        blast_genus = sub[
            (sub["avg_pid"] >= GENUS_PID) &
            (sub["avg_qcov"] >= GENUS_QCOV) &
            (sub["blast_genus"] != "")
        ]

        if not blast_genus.empty:
            best = blast_genus.sort_values("avg_pid", ascending=False).iloc[0]
            row.update({
                "rule": "B",
                "method": "BLAST_genus",
                "genus": best["blast_genus"],
                "species": "",
                "confidence": best["avg_pid"]
            })
            assignments.append(row)
            continue

        # ---------- Rule C: SINTAX ----------
        srow = sub.iloc[0]
        if srow["sintax_species_score"] >= SINTAX_SPECIES_CUTOFF:
            row.update({
                "rule": "C",
                "method": "SINTAX_species",
                "genus": srow["sintax_genus"],
                "species": srow["sintax_species"],
                "confidence": srow["sintax_species_score"]
            })
            assignments.append(row)
            continue

        if srow["sintax_genus_score"] >= SINTAX_GENUS_CUTOFF:
            row.update({
                "rule": "C",
                "method": "SINTAX_genus",
                "genus": srow["sintax_genus"],
                "species": "",
                "confidence": srow["sintax_genus_score"]
            })
            assignments.append(row)
            continue

        # ---------- Fallback ----------
        row.update({
            "rule": "Fallback",
            "method": "Fungi_sp",
            "genus": "Fungi",
            "species": "",
            "confidence": 0.0
        })
        assignments.append(row)

    return pd.DataFrame(assignments)

# ================= MERGE RESULTS =================
def combine_and_assign(blast_df, sintax_df):
    merged = blast_df.merge(
        sintax_df,
        on="qseqid",
        how="left"
    )
    final = assign_taxonomy(merged)
    return merged, final

# ================= PLOTS =================
def plot_histograms(df):
    plt.figure()
    df["avg_pid"].dropna().hist(bins=40)
    plt.xlabel("Average % identity")
    plt.ylabel("Count")
    plt.title("BLAST average identity distribution")
    plt.savefig(os.path.join(PLOTS_DIR, "blast_avg_pid.png"))
    plt.close()

    plt.figure()
    df["avg_qcov"].dropna().hist(bins=40)
    plt.xlabel("Average query coverage (%)")
    plt.ylabel("Count")
    plt.title("BLAST average query coverage")
    plt.savefig(os.path.join(PLOTS_DIR, "blast_avg_qcov.png"))
    plt.close()

# ================= MAIN =================
def main():
    logger.info("🚀 Starting hybrid taxonomic pipeline")

    gather_queries()

    n_otus = sum(1 for _ in SeqIO.parse(OTU_QUERIES, "fasta"))
    parallel = n_otus > PARALLEL_THRESHOLD

    run_blast(parallel=parallel)
    run_sintax(parallel=parallel)

    acc_map = build_unite_accession_map(UNITE_FASTA)

    blast_df = parse_blast(BLAST_OUT, acc_map)
    sintax_df = parse_sintax(SINTAX_OUT)

    merged, final = combine_and_assign(blast_df, sintax_df)

    merged.to_csv(FINAL_VERBOSE, sep="\t", index=False)
    final.to_csv(FINAL_ONE, sep="\t", index=False)

    unassigned = final[final["method"] == "Fungi_sp"]
    if not unassigned.empty:
        unassigned.to_csv(UNASSIGNED, sep="\t", index=False)

    plot_histograms(merged)

    logger.info("✅ Pipeline completed successfully")
    logger.info("📄 Final taxonomy: %s", FINAL_ONE)

# ================= ENTRY =================
if __name__ == "__main__":
    main()
```
